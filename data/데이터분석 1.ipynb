```python
df.loc[:,'year'] #year칼럼의 모든 행
df.loc[1:4,'year'] #year칼럼의 1~4
df.loc[:4,'year'] #year칼럼의 0~4
df.loc[1700:,'year'] #year칼럼의 1700~끝
```

```python
df.loc[1,:] #1번행 모든 칼럼
df.loc[1,: 'pop'] #1번행 pop열까지
df.loc[1,'year' : 'pop'] #1번행 year열부터 pop열까지
df.loc[1,'year' : ] #1번행 year열부터 마지막까지
df.loc[1:4, 'year' : 'pop'] #1~4행 year열부터 pop까지
df.loc[[4,1,2], 'year' : 'pop'] #4,1,2번행, year열부터 pop열까지
```

```python
#df.loc[조건식] #조건식을 만족하는 행 추출
b1 = df['year'] < 1960
df.loc[b1] #True인 행만 추출
df.loc[df['year'] <1960 ] #한줄로
```
