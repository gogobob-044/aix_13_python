```python
import seaborn as sns
import pandas as pd

tips = sns.load_dataset('tips')
tips.info()
```

```python
tips.head()
```

```python
tips.head()
```

```python
import matplotlib.pyplot as plt
plt.hist(data=tips, x='total_bill',bins=20) #bins:구간경계값개수
#한글폰트설정
plt.rc('font',family='Malgun Gothic')
plt.title("팁")
plt.show()
```

```python
plt.scatter(
    data=tips,
    x='total_bill',
    y='tip'
)
#요금을 기준으로 팁을 봄
plt.title("팁")
plt.xlabel("요금")
plt.ylabel("팁")
plt.show()
```

```python
plt.scatter(
    data=tips.query("sex=='Male'"),
    x="total_bill",
    y="tip"
)
plt.scatter(
    data=tips.query("sex=='Female'"),
    x="total_bill",
    y="tip"
)
plt.legend() # 범례
```

```python
# 종류별로
# 성별
plt.scatter(
    data=tips.query("sex=='Male'"),
    x="total_bill",
    y="tip",
    label="Male"
)
plt.scatter(
    data=tips.query("sex=='Female'"),
    x="total_bill",
    y="tip",
    label="Female"#맨위 라벨
)
plt.legend() # 범례
```

```python
plt.boxplot(
    [tips.query('sex=="Male"')['tip'],
    tips.query('sex =="Female"')['tip'],], #콤마 잊지말기~
    labels=['Male','Female'] #리스트로 받아서 라이브러리를 복수형으로 지정함
)
plt.show()
```

```python
tips.query('sex=="Male"')['total_bill'].plot.hist()
tips.query('sex=="Female"')['total_bill'].plot.hist()
```

```python
tips.query('sex=="Male"')['total_bill'].plot.hist(alpha=.5) #투명도
tips.query('sex=="Female"')['total_bill'].plot.hist(alpha=.5)
```

```python
ax = tips.query('sex=="Male"').plot.scatter( x='total_bill', y='tip', c='red')
tips.query('sex=="Female"').plot.scatter( x='total_bill', y='tip', ax=ax) #그린공간에 다시 그린다는 뜻

#2개 겹쳐 seaborn쓰는게 젤 편함
```
