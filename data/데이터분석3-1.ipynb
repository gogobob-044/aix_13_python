```python
df1.columns = ['A', 'B', 'C', 'D'] #기존에 있는 열 이름만 바꾸는거
df2.columns = ['E', 'F', 'G', 'H']
df3.columns = ['A', 'C', 'F', 'H']
print(df1)
print(df2)
print(df3)
```

```python
row_concat = pd.concat([df1, df2, df3])
print(row_concat)
```

```python
#공통부분만 남김
row_concat = pd.concat([df1, df3],join='inner')
print(row_concat)
```

```python
df1.index = [0, 1, 2, 3]
df2.index = [4, 5, 6, 7]
df3.index = [0, 2, 5, 7]
print(df1)
print(df2)
print(df3)
```

```python
col_concat = pd.concat([df1,df2, df3], axis="columns",join='outer')
print(col_concat)
```

```python
col_concat = pd.concat([df1, df3], axis="columns",join='outer')
print(col_concat)
```

```python
col_concat = pd.concat([df1, df3], axis="columns",join='inner')
print(col_concat)
#join='inner' 공통 부분만
```

```python
person = pd.read_csv('./data/survey_person.csv')
site = pd.read_csv('./data/survey_site.csv')
survey = pd.read_csv('./data/survey_survey.csv')
visited = pd.read_csv('./data/survey_visited.csv')
print(person)
print(survey)
```

```python
visited_subset = visited.loc[[0, 2, 6], :]
print(visited_subset)
```

```python
#인덱스말고 값으로 연결
o2o_merge = site.merge(visited_subset, left_on="name", right_on="site")
print(o2o_merge)
```

```python
#6이 빠지면
visited_subset = visited.loc[[0, 2,], :]
print(visited_subset)
```

```python
o2o_merge = site.merge(visited_subset, left_on="name", right_on="site", how='left')
print(o2o_merge)
```
