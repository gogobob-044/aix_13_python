```python
# 성별 나이평균 계산하기
# 전테탑승객중 최고나이, 최소나이
# 성별 최고나이,최소나이
age_Max = df['Age'].max()
age_Min = df['Age'].min()
print(age_Max, age_Min)

df.groupby('Sex')['Age'].max()
df.groupby('Sex')['Age'].min()
#문제출제~
```

```python
#선실 등급별 최고,최소,평균

df.groupby('Pclass')['Age'].max()
df.groupby('Pclass')['Age'].min()
df.groupby('Pclass')['Age'].mean()
```

```python
#성별 생존율(Surviveddml 평균)
df.groupby('Sex')['Survived'].mean()
#선실 등급별 생존율
df.groupby('Pclass')['Survived'].mean()
#성별, 선실 등급별 생존율
df.groupby(['Sex','Pclass'])['Survived'].mean()
```

```python
#Age의 결측치 채우기(평균)
b1=df['Age'].isna()
df[b1] #Age가 없는 값만 추출
```

```python
age_mean = df['Age'].mean()
df['Age'].fillna(age_mean) #주의!! 채운 df복사본이 반환. 원본 수정 안됨
#원본수정방법
#1.inplace = True 부여 #원본을 불러온다 #잘못해도 끝임
#2.원본칼럼을 덮어쓴다. 
df['Age'] = df['Age'].fillna(age_mean) 
df
```
