```python
import re
tel_num = "0101112222 gildon@abc.com"
m = re.match(pattern = '\d\d\d\d\d\d\d\d\d\d', string=tel_num)
print(type(m))
print(m)
```

```python
tel_num = "홍길동 010-111-2222 gildon@abc.com"

# m = re.search(pattern = '\d\d\d\d\d\d\d\d\d\d', string=tel_num)
# m = re.search(pattern = '\d{10}', string=tel_num)
m = re.search(pattern = '\d{3}-\d{3}-\d{4}', string=tel_num)
print(type(m))
print(m)
```

```python

# tele_num_space_paren_dash = '(123) 456-7890'
tele_num_space_paren_dash = '123)456 7890'
p = '\(?\d{3}\)?\s?\d{3}\s?-?\d{4}'  #\(? 글자(가 있거나없거나
#\s? 공백이 있거나 없거나 -> 생략 가능하다는 뜻!!
m = re.match(pattern=p, string=tele_num_space_paren_dash)
print(m)
```
