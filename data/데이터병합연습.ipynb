```python
import pandas as pd

customers = pd.DataFrame({
    'customer_id': [1, 2, 3, 4],
    'name': ['Alice', 'Bob', 'Charlie', 'David']
})

orders = pd.DataFrame({
    'order_id': [101, 102, 103],
    'customer_id': [1, 2, 5],
    'amount': [25000, 45000, 30000]
})
print(customers)
print(orders)
```

```python
#왜 ㅏㅇㄴ돼
#co_merge = customers.merge(orders, left_on="customer_id",right_on='order_id')

# 방법 1: 공통 컬럼 이름이 같을 때 가장 간단한 방법
co_merge = customers.merge(orders, on='customer_id')
print(co_merge)
```

```python
co_merge = customers.merge(orders, on='customer_id', how='left') #왼쪽기준으로 자르기
print(co_merge)
```

```python
co_merge = customers.merge(orders, on='customer_id', how='right') #오른쪽기준으로 자르기
print(co_merge)
```
