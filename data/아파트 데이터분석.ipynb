```python
#평을 평형으로
def to_ph(x):
    if x < 10: return "10평 미만"
    if x < 20: return "10평대"
    if x < 30: return "20평대"
    if x < 40: return "30평대"
    return "40평 이상"

df['평형']=df['평'].apply(to_ph)
```

```python
#평형별 평균 거래금액
df.groupby('평형')['거래금액(만원)'].mean().sort_values(ascending=False)
```

```python
# 평형별 최고평
df.groupby('평형')['평'].max().sort_values(ascending=False)
```

```python
#구별 평형별 거래건수
df.groupby(['구','평형'])['층'].sum()
```

```python
#전체 평균 거래금액에 가장 가까운 아파트를 찾으세요

# 1. 전체 평균 거래금액을 a에 저장
a = df['거래금액(만원)'].mean()

# 2. 거래금액과 a간의 차이값(절대값)을 diff 컬럼에 저장
# 평균보다 적거나 많을 수 있으므로 abs()
df['diff'] = (df['거래금액(만원)'] - a).abs()

# 3. diff 컬럼 기준 오름차순 정렬 후 맨 처음 행 추출
closest_apt = df.sort_values(by='diff').iloc[0]

print(f'전체 평균 거래금액:{a}원')
print("평균에 가장 가까운 아파트 정보:")
print(closest_apt[['단지명', '전용면적(㎡)', '거래금액(만원)', 'diff']])
```

```python
#계약년월 계약일 ->요일칼럼추가

# 2. 날짜 형식으로 변환
date = pd.to_datetime(df['계약년월'].astype(str) + 
    df['계약일'].astype(int).astype(str).str.zfill(2), format='%Y%m%d')

df['요일'] = date.dt.day_name()
weekday_map = {
    'Monday': '월', 'Tuesday': '화', 'Wednesday': '수', 
    'Thursday': '목', 'Friday': '금', 'Saturday': '토', 'Sunday': '일'
}
df['요일'] = df['요일'].map(weekday_map)

# 결과 확인
print(df[['계약년월', '계약일', '요일']].head())
```

```python
# 요일별로 묶어서(groupby) 거래금액의 개수(count)와 평균(mean) 구하기
count = df.groupby('요일')['층'].count()
mean = df.groupby('요일')['거래금액(만원)'].mean()
print(count)
print(mean)
```
